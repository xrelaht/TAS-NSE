```python
import numpy as np
import os
import lmfit 
import matplotlib.pyplot as plt
from lmfit import minimize, Parameters, Parameter, report_fit 
from matplotlib.backends.backend_pdf import PdfPages
from IPython.display import Image
import re
%matplotlib inline
```

```python
def cosfit(p, x, y):
    return p['I0'].value/2.0 *(1.0 + p['Pol'].value * np.cos(2.0*np.pi*(x-p['x0'].value)/p['dL'].value))-y

def get_line(parameter,fileno,path='/Users/steven/Dropbox/Data 2017/051517 TRISP Ge/data_May_16_a/'):
    os.chdir(path)
    i=0
    with open('sc'+str(fileno)+'.log') as f:
            for line in f:
                if parameter in line:
                      return line
                i=i+1
def get_para(parameter,fileno,path):
    outstr=get_line(parameter,fileno,path)
    numb=re.findall("[-+]?\d*\.\d+|\d+",outstr)
    return np.float(numb[0])

class trisp(object):
    def __init__(self,path, runno,xcol=2, ycol=8, backround=0):
        current_path=os.getcwd()
        os.chdir(path)
        data=np.loadtxt('sc'+str(runno)+'.dat',skiprows=1)
        self.x = data[:,xcol]
        self.y = data[:,ycol]
        self.e = np.sqrt(self.y)
        self.P0_s=(self.y.max()-self.y.min())/(2*self.y.mean())
        self.kf=np.genfromtxt('sc'+str(runno)+'.log',skiprows=56,usecols=1,skip_footer=246-56)
        self.RFM4COILS=np.genfromtxt('sc'+str(runno)+'.log',skiprows=63,usecols=1,skip_footer=247-64)
        self.I0_s=2.0*self.y.mean()
        self.x0_s=max(xrange(len(self.y)), key = lambda x: self.y[x])
        self.RFMODE=4.0
        self.FREQ1=data[0,4]
        self.FREQ2=data[0,5]
        self.dL_s=dL_s = 3956/(2*np.pi)* self.kf / (self.FREQ2 * self.RFMODE / 2.0);
        p= lmfit.Parameters()
        p.add_many(('Pol', self.P0_s, True, -1.0, 1.0), ('I0', self.I0_s, True), ('x0', self.x0_s, True),('dL',self.dL_s,True))
        self.kkds=lmfit.minimize(cosfit, p, args=(self.x,self.y)) 
        self.pol=np.abs(self.kkds.params['Pol'].value)
        self.I0 =self.kkds.params['I0'].value 
        self.x0 =self.kkds.params['x0'].value 
        self.dl =self.kkds.params['dL'].value 
        os.chdir(current_path)
        
    def plot(self, **kwargs):
        plt.errorbar(self.x, self.y, yerr=self.e, **kwargs)
        plt.ylabel('counts', fontsize=10)
        plt.xlabel('tau[s$', fontsize=10)
        plt.plot(self.x, self.kkds.residual+self.y)
```

```python
data_path='/Users/steven/Dropbox/Data 2017/051517 TRISP Ge/data_May_16_a/'

runmin=85644
runmax=85653
x=np.zeros(10)
y=np.zeros(10)
tau=np.zeros(10)
j=0
for i in range (runmin,runmax+1):
    fitdata=trisp(data_path,i)
    x[j]=j
    tau[j]=  4.087 * fitdata.FREQ2 / 50
    y[j]=fitdata.pol
    j=j+1
    print(fitdata.RFM4COILS)
    plt.figure()
    fitdata.plot()
#    plt.figure()
#plt.plot(tau,np.abs(y),'o')
```
